```python
import pandas as pd
df = pd.read_csv('netflix_titles.csv')

print(df.head(10))
```

```python
print(df['type'].value_counts().plot(kind='bar',title='No. Of Content By Type'))
```

```python
print(df[df['year_added'].between(2015,2021)]['year_added'].value_counts().plot(kind='bar',title='Titles Added Per Year By Netflix'))
```

```python
print(df['type'].value_counts())
```

```python
print(df['country'].value_counts().head(10))
df['country'].value_counts().head(10).plot(kind='bar',title='Top 10 Countries on Netflix',)
```

```python
print(df['release_year'].value_counts().sort_values())
df['release_year'].value_counts().sort_values().plot(kind='line',figsize=(10,4),title='Netflix releases over time')
```

```python
# Most common Genre
print(df['listed_in'].value_counts().head(10))
```

```python
from collections import Counter
genres = df['listed_in'].dropna().str.split(',')
flat_genres = [genre for sublist in genres for genre in sublist]
top_genres = pd.Series(Counter(flat_genres)).sort_values(ascending = False).head(10)
print(top_genres)
```
